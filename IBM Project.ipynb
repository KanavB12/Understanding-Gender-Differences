```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 150)
```

```python
young = pd.read_csv(r'D:\Users\DELL\Desktop\Padai\Sem 6\Data Mining\responses.csv')
young.head(2)
```

```python
young.describe()
```

```python
var_of_interest = 'Gender'
mapping = {var_of_interest: {'male':0, 'female':1}}
young.dropna(subset=[var_of_interest], inplace=True)
```

```python
df = pd.get_dummies(young, columns=["Smoking", "Alcohol", "Punctuality",
                                   "Lying", "Internet usage", "Left - right handed",
                                   "Education", "Only child", "Village - town", "House - block of flats"])
```

```python
cols = list(df.columns)
cols.remove("Gender")
```

```python
fig, ax = plt.subplots(nrows=24, ncols=8, figsize=(30, 70), sharex=True)
start=0
for j in range(24):
    for i in range(8):
        sns.barplot(
        y=cols[start], x=var_of_interest, data=df, ax=ax[j, i])
        ax[j, i].set_ylabel('')
        ax[j, i].set_xlabel('')
        ax[j, i].set_title(cols[start], fontsize=25)
        start += 1
```

```python
import scipy.stats as stats
data = df.dropna(subset=['Reading'])
male_read = data[data[var_of_interest] == 'male']['Reading']
female_read = data[data[var_of_interest] == 'female']['Reading']
t, p = stats.ttest_ind(male_read, female_read, axis=0, equal_var=False)
print('t-stat = {t} \n p-value = {p}'.format(t=t, p=p/2))
```

```python
good_columns = []
count = 0
for col in cols:
    data = df.dropna(subset=[col])
    male = data[data[var_of_interest] == 'male'][col]
    female = data[data[var_of_interest] == 'female'][col]
    t, p = stats.ttest_ind(male, female, axis=0, equal_var=False)
    if(p < 0.01):
        print(col + " : p-value = {p}".format(p=p/2))
        count += 1
        good_columns.append(col)
```

```python
print(count)
good_columns
```

```python
data = df[good_columns]
data.head(2)
```

```python
corr = data.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(corr)
```

```python
remove_columns = []
for i in range(0, corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if abs(corr.iloc[i, j]) >= 0.7:
            col = good_columns[j]
            if col not in remove_columns:
                remove_columns.append(col)
```

```python
remove_columns
```

```python
df2 = data.drop(remove_columns, axis=1)
len(df2.columns)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
```

```python
X = df2
mean_values = X.mean(axis=0)
X = X.apply(lambda x: x.fillna(x.mean()),axis=0)

gender = young['Gender']
y = pd.get_dummies(gender)
y = y["female"]
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size=0.2)
print(xTrain.shape, xTest.shape, yTrain.shape, yTest.shape)
```

```python
# standardization
scaler = StandardScaler()
scaler.fit(xTrain)
xTrain = scaler.transform(xTrain)
xTest = scaler.transform(xTest)
```

```python
from sklearn.metrics import recall_score, precision_score
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score
```

```python
# Grid Search for parameter tuning
clr = LogisticRegression()
KF = KFold(len(xTrain), n_splits=5)
```

### Logistic Regression

```python
param_grid = {'C':[.001, .01, .03, .1, .3, 1, 3, 10]}
grsearch = GridSearchCV(clr, param_grid=param_grid, cv=KF, scoring='f1')
grsearch.fit(xTrain, yTrain)
print(grsearch.best_params_)
```

```python
# fitting logistic regression and evaluating
clr = LogisticRegression(C=grsearch.best_params_['C'])
clr.fit(xTrain, yTrain)

mean_accuracy = np.mean(cross_val_score(clr, xTrain, yTrain, cv=KF))
print('Average accuracy score on CV set: {:.2f}'.format(mean_accuracy))

yPred = clr.predict(xTest)
print(f1_score(yTest, yPred))
print(confusion_matrix(yTest, yPred))
print(classification_report(yTest, yPred))
```

```python
feat_coeff = pd.DataFrame({'impacts': clr.coef_[0], 'features': X.columns})
feat_coeff = feat_coeff.sort_values('impacts', ascending=False)

sns.set_style("whitegrid")
fig, ax1 = plt.subplots(1,1, figsize=(6,30));
sns.barplot(x=feat_coeff.impacts, y=feat_coeff.features, ax=ax1, palette='deep');
# ax1.set_title('All features', size=30); 
# ax1.set_xticklabels(labels=feat_coeff.impacts, size=20, rotation=90);
# ax1.set_ylabel('Impact', size=30);
```

```python
top10 = pd.concat([feat_coeff.head(5),feat_coeff.tail(5)])
fig, ax1 = plt.subplots(1,1, figsize=(10,6))
sns.barplot(y=top10.features, x=top10.impacts, ax=ax1);
ax1.set_title('Top 10 features', size=20);
ax1.set_yticklabels(labels=top10.features, size=15);
ax1.set_xlabel('Impact', size=20);
```

### K-Nearest Neighbors

```python
clr_knn = KNeighborsClassifier()
param_grid = {'n_neighbors':[3, 4, 5, 6, 7, 8, 9, 10]}
grsearch = GridSearchCV(clr_knn, param_grid=param_grid, cv=KF, scoring='f1')
grsearch.fit(xTrain, yTrain)
print(grsearch.best_params_)
```

```python
# fitting KNN and evaluating
clr_knn = KNeighborsClassifier(n_neighbors=grsearch.best_params_['n_neighbors'])
clr_knn.fit(xTrain, yTrain)

mean_accuracy = np.mean(cross_val_score(clr_knn, xTrain, yTrain, cv=KF))
print('Average accuracy score on CV set: {:.2f}'.format(mean_accuracy))

yPred = clr_knn.predict(xTest)
print(f1_score(yTest, yPred))
print(confusion_matrix(yTest, yPred))
print(classification_report(yTest, yPred))
```

### Support Vector Classifier

```python
clr_svc = svm.SVC()
param_grid = {'C':[.001, .01, .1, 1, 3], 
             'kernel':['linear', 'poly', 'rbf', 'sigmoid']}
grsearch = GridSearchCV(clr_svc, param_grid=param_grid, cv=KF, scoring='f1')
grsearch.fit(xTrain, yTrain)
print(grsearch.best_params_)
```

```python
# fitting SVC and evaluating
clr_svc = svm.SVC(C=grsearch.best_params_['C'], kernel=grsearch.best_params_['kernel'])
clr_svc.fit(xTrain, yTrain)

mean_accuracy = np.mean(cross_val_score(clr_svc, xTrain, yTrain, cv=KF))
print('Average accuracy score on CV set: {:.2f}'.format(mean_accuracy))

yPred = clr_svc.predict(xTest)
print(f1_score(yTest, yPred))
print(confusion_matrix(yTest, yPred))
print(classification_report(yTest, yPred))
```
